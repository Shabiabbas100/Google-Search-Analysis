### Script for analysing the Google search trends on the queries based on "Grok AI". Here, a dataframe of the top 10 countries with the highest interest is considered.

```python
pytrends.build_payload([keyword], timeframe="2023-12-01 2025-04-30", geo="") #making a request

df = pytrends.interest_by_region()  # to find the popularity in the region worldwide
df=df.sort_values(by=keyword,ascending =False)  
data=df.head(10)

print(data)     # just top 10 countries
```

<br><br>
NOTE: The values are calculated on a scale from 0 to 100, where 100 is the location with the most popularity as a fraction of total searches in that location, a value of 50 indicates a location which is half as popular. A value of 0 indicates a location where there was not enough data for this term.

<br> <br>

### From the above results, **India** has the highest search query in **Grok AI**

### Visualisation of data using bar chart

```python
data.reset_index().plot(x="geoName", y=keyword, figsize=(16,12), kind="bar")
plt.style.use('tableau-colorblind10')
plt.show()
```

<br><br>
### We know that Grok, the AI chatbot from xAI, has gained a lot of attention recently, so here’s a look at the trend of searches over the past few years.

```python
data =pytrends.interest_over_time()
print(data)
fig =plt.figure(figsize=(15,20));
axis=fig.add_axes([0.1,0.1,1,1])
axis.plot(data.index,data[keyword],linewidth=3)
plt.title(f'Total Google Searches for {keyword}', fontweight='bold')
plt.style.use('fivethirtyeight')
plt.xlabel('Year',fontweight='bold')
plt.ylabel('Total Count',fontweight='bold')
```
